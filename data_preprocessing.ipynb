```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
from glob import glob as gl
from tqdm import tqdm, tnrange
```

```python
RAW_VIDEO_DATA_DIR = 'data/raw_video/other'
RAW_VIDEO_DATA_PATH = gl(RAW_VIDEO_DATA_DIR + '/*')

EXTRACTED_FRAME_DIR = 'data/extracted_frame/other'
```

```python
RAW_VIDEO_DATA_PATH
```

```python
def get_frame(cap, sec, video_name, frame_index):
    cap.set(cv2.CAP_PROP_POS_MSEC, sec * 1000)
    success, image = cap.read()
    if success:
        cv2.imwrite(EXTRACTED_FRAME_DIR + '/img_{}_{}.jpg'.format(video_name, frame_index), image)
    
    return success

total_frame = 0
for vid_path in RAW_VIDEO_DATA_PATH:
    cap = cv2.VideoCapture(vid_path)
    video_name = vid_path.split('/')[-1].split('.mp4')[0].split('\\')[-1]
    frame_index = 0
    frame_rate = 7
    sec = 0
    
    success = get_frame(cap, sec, video_name, frame_index)
    while success:
        frame_index += 1
        total_frame += 1
        sec = sec + frame_rate
        sec = round(sec, 4)
        success = get_frame(cap, sec, video_name, frame_index)
    print('Done ', video_name)
    cap.release()
```
